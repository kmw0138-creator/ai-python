```python
# 동일한 키('아메리카노')가 중복 입력된 상황
prices = {'아메리카노': 2000, '라떼': 3000, '아메리카노': 1900, '녹차': 2500}
print(prices)
# 결과: {'아메리카노': 1900, '라떼': 3000, '녹차': 2500}
# 2000원이었던 아메리카노 값이 마지막 값인 1900으로 덮어씌워졌습니다.
```

```python
# pop() 함수 활용 예시
prices = {'아메리카노': 2000, '라떼': 3000, '녹차': 2500}
a = prices.pop('라떼') # '라떼' 항목을 삭제하고, 가격인 3000을 a에 저장
print(a) # 출력: 3000
print(prices) # 출력: {'아메리카노': 2000, '녹차': 2500}
```

```python
# 1. keys() 활용: 이름표만 확인
for key in prices.keys():
    print(key, end=' ') # 아메리카노 라떼 녹차
# 2. values() 활용: 가격들만 확인
for val in prices.values():
    print(val) # 2000 \n 3000 \n 2500
# 3. items() 활용: 이름과 가격을 동시에!
for key, value in prices.items():
    print(f"{key}의 가격은 {value}원입니다.")
```
