```python
# 1. 사용자로부터 숫자 입력받기
user_input = input("숫자를 입력하세요: ")

# 2. 입력받은 문자를 정수형(int) 숫자로 변환
number = int(user_input)

# 3. 짝수와 홀수 판별하기
if number % 2 == 0:
    print(f"{number}은(는) 짝수입니다.")
else:
    print(f"{number}은(는) 홀수입니다.")
```

```python
# 1. 사용자로부터 숫자 입력받기
user_input = input("숫자를 입력하세요: ")

# 2. 입력받은 문자를 정수형(int) 숫자로 변환
number = int(user_input)

# 3. 짝수와 홀수 판별하기
if number > 0 :
    print(f"{number}은(는) 양수입니다.")
else:
    print(f"{number}은(는) 음수입니다.")
```

```python
cp= 2000
lp= 3000
gp= 2500
jp= 2200

cn = int(input('커피 몇 잔?: '))
ln = int(input('라떼 몇 잔?: '))
gn = int(input('녹차 몇 잔?: '))
jn = int(input('주스 몇 잔?: '))

total = int(cp*cn + lp*ln +gp*gn + jp*jn)
print(f"{total} 내놔.")
```
