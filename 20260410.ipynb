```python
kings = ["태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종", "연산군"]
```

```python
for I in list(reversed(kings)):
    print(I)
```

```python
order=int(input("몇 번째 왕을 알고 싶으신가요?(1~10):"))
if 1<=order<=len(kings):
    kings_name = kings[order-1]
    print(f"조선의 {order}대 왕은 {kings_name}입니다")
else:
    print(f"죄송합니다 현재는 {len(kings)}대 왕까지만 저장되어있습니다")
```
